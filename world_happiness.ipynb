```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# load the dataset
dataset=pd.read_csv("world-happiness-report.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.rename(columns={'Country name':'Country',
                     'Life Ladder':'Life_ladder',
                     'Log GDP per capita':'Log_GDP_per_capita',
                     'Social support' : 'Social_support',
                     'Healthy life expectancy at birth':'Healthy_life_expectancy_at_birth',
                     'Freedom to make life choices':'Freedom_to_make_life_choices',
                     'Perceptions of corruption' : 'Perceptions_of_corruption',
                     'Positive affect' : 'Positive_affect',
                     'Negative affect': 'Negative_affect'},inplace=True)
```

```python
dataset.columns
```

```python
# filling the missing values
dataset['Log_GDP_per_capita'] = dataset['Log_GDP_per_capita'].fillna(np.round(dataset['Log_GDP_per_capita'].mean()))
dataset['Social_support'] = dataset['Social_support'].fillna(np.round(dataset['Social_support'].mean()))
dataset['Healthy_life_expectancy_at_birth'] = dataset['Healthy_life_expectancy_at_birth'].fillna(np.round(dataset['Healthy_life_expectancy_at_birth'].mean()))
dataset['Freedom_to_make_life_choices'] = dataset['Freedom_to_make_life_choices'].fillna(np.round(dataset['Freedom_to_make_life_choices'].mean()))
dataset['Generosity'] = dataset['Generosity'].fillna(np.round(dataset['Generosity'].mean()))
dataset['Perceptions_of_corruption'] = dataset['Perceptions_of_corruption'].fillna(np.round(dataset['Perceptions_of_corruption'].mean()))
dataset['Positive_affect'] = dataset['Positive_affect'].fillna(np.round(dataset['Positive_affect'].mean()))
dataset['Negative_affect'] = dataset['Negative_affect'].fillna(np.round(dataset['Negative_affect'].mean()))
```

```python
# Let's draw a histogram chart to see dataset frequency information
def plotHistogram(column):
  """
  Input: Column name
  Output: histogram chart of the relevant feature
  """
  plt.figure()
  plt.hist(dataset[column],bins=85,color="orange")
  plt.xlabel(column)
  plt.ylabel("Frequency")
  plt.title("dataset Frequency - {}".format(column))
  plt.show()
```

```python
# histogram charts for all numeric variables
numeric_values=['year', 'Life_ladder', 'Log_GDP_per_capita',
       'Social_support', 'Healthy_life_expectancy_at_birth',
       'Freedom_to_make_life_choices', 'Generosity',
       'Perceptions_of_corruption', 'Positive_affect', 'Negative_affect']
for i in numeric_values:
  plotHistogram(i)
```

```python
def plotBox(column):
  plt.figure()
  plt.boxplot(dataset[column])
  plt.xlabel(column)
  plt.ylabel("Value")
  plt.show()
```

```python
# Let's draw a box plot to see outliers
columns=["Life_ladder","Social_support","Freedom_to_make_life_choices","Generosity","Positive_affect","Negative_affect"]
for i in columns:
  plotBox(i)
```

```python
# let's draw bar charts to examine categorical variables
def plotBar(column,n=5):
  """
  Input: column name
        n= Number of unique values ​​to display
  Output : Bar chart
  """
  dataset_ = dataset[column]
  count_dataset=dataset_.value_counts()
  count_dataset= count_dataset[:n]

  plt.figure()
  plt.bar(count_dataset.index,count_dataset,color="orange")
  plt.xticks(count_dataset.index,rotation=20)
  plt.ylabel("Frequency")
  plt.title("dataset Frequency - {}".format(column))
  plt.show()
  print("{} : \n {}".format(column,count_dataset))
```

```python
# anomaly detection
# outlier analysis
from collections import Counter
def anomaliDetection(df,column):
  outlier_indices = []
  for c in column:
    # 1st quarter
    Q1= np.percentile(df[c],25)
    # 3rd quarter
    Q3=np.percentile(df[c],75)
    # IQR = Inter Quartile Range
    IQR=Q3-Q1
    # additional step amount for outlier
    outlier_step = 1.5 * IQR
    # Let's detect the outlier value and the index it is found in.
    outlier_list_col=df[(df[c]< Q1-outlier_step) | (df[c] > Q3 + outlier_step)].index
    # Let's store the detected indexes
    outlier_indices.extend(outlier_list_col)


  # Let's find unique outliers
  outlier_indices = Counter(outlier_indices)
  # If an example has different values ​​in v columns, let's consider it as an outlier
  multiple_outliers=list(i for i,v in outlier_indices.items() if v>1)

  return multiple_outliers
```

```python
data_anomali=dataset.loc[anomaliDetection(dataset,['year', 'Life_ladder', 'Log_GDP_per_capita',
       'Social_support', 'Healthy_life_expectancy_at_birth',
       'Freedom_to_make_life_choices', 'Generosity',
       'Perceptions_of_corruption', 'Positive_affect', 'Negative_affect'])]
data_anomali.Country.value_counts()
```

```python
# data analysis in time series

columns=['Life_ladder', 'Log_GDP_per_capita','Social_support', 'Healthy_life_expectancy_at_birth','Freedom_to_make_life_choices', 'Generosity',
       'Perceptions_of_corruption','Positive_affect', 'Negative_affect']
dataset['year'] = pd.to_datetime(dataset['year'], format='%Y')
dataset.set_index('year', inplace=True)


```

```python

for col in columns:
    plt.scatter(dataset.index,dataset[col])
    plt.title(f'Time Series of {col}')
    plt.xlabel('Year')
    plt.ylabel(col)
    plt.show()
```

```python
# selecting the top 3 countries according to Life Ladder value
top_countries = dataset.groupby('Country')['Life_ladder'].mean().nlargest(3).index
for country in top_countries:
    country_data = dataset[dataset['Country'] == country]
    country_data[columns].plot(subplots=True, layout=(3, 3), figsize=(15, 10), title=country)
    plt.tight_layout()
    plt.show()
```
